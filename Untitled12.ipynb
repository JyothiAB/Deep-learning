```python
#Dataset for bank customer churn prediction : https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/JyothiAB/Deep-learning/1_Bank-Customer-Churn-Prediction/Churn_Modelling.csv')
df.head()
```

```python
df.drop(columns=['RowNumber','Surname','CustomerId'], inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']] = scaler.fit_transform(df[['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']])
```

```python
df.describe()
```

```python
df['Gender'].replace({'Male':0,'Female':1}, inplace=True)
```

```python
df['Gender'].describe()
```

```python
df1 = pd.get_dummies(data = df, columns=['Geography'])
```

```python
df1.describe()
```

```python
X = df1.drop(columns = ['Exited'])
y = df1['Exited']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=5)
```

```python
len(x_train.columns)
```

```python
def ann_model(x_train,x_test,y_train,y_test):

  import tensorflow as tf
  from tensorflow import keras

  model = keras.Sequential([
                          keras.layers.Dense(12, input_shape = (12,), activation='relu'),
                          keras.layers.Dense(1, activation='sigmoid')
                          ])
  
  model.compile(optimizer='adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
  
  model.fit(x_train,y_train,epochs = 100)

  y_pred = model.predict(x_test)
  y_predict = []
  for i in y_pred:
    if i < 0.5:
      y_predict.append(0)
    else:
      y_predict.append(1)

  from sklearn.metrics import confusion_matrix , classification_report

  print(classification_report(y_test,y_predict))
  return y_predict
```

```python
ann_model(x_train,x_test,y_train,y_test)
```

```python
#Method 1: Undersampling the Majority class
df_train = pd.concat([x_train,y_train], axis = 'columns')
df_train.columns
```

```python
df_train_undersampling = pd.concat([df_train[df_train['Exited'] == 0].sample(1632), df_train[df_train['Exited'] == 1]], axis = 'rows')
x_train_undersampling = df_train_undersampling.drop(columns = 'Exited')
y_train_undersampling = df_train_undersampling['Exited']
```

```python
y_train_undersampling.value_counts()
```

```python
ann_model(x_train_undersampling,x_test,y_train_undersampling,y_test)
```

```python
#Method 2: Random oversampling the minority class
```

```python
df_train_oversampling = pd.concat([df_train[df_train['Exited'] == 0], df_train[df_train['Exited'] == 1].sample(6368, replace=True)], axis = 'rows')
x_train_oversampling = df_train_oversampling.drop(columns = 'Exited')
y_train_oversampling = df_train_oversampling['Exited']
```

```python
y_train_oversampling.value_counts()
```

```python
ann_model(x_train_oversampling,x_test,y_train_oversampling,y_test)
```

```python
#Method 3: SMOTE

from imblearn.over_sampling import SMOTE
smote_data = SMOTE(sampling_strategy='minority', k_neighbors = 5)
x_train_smote,y_train_smote = smote_data.fit_resample(x_train,y_train)
```

```python
y_train_smote.value_counts()
```

```python
ann_model(x_train_smote,x_test,y_train_smote,y_test)
```

```python
# Method 4 Ensemble technique
6368/1632
df_train_0 = df_train[df_train['Exited']==0]
df_train_1 = df_train[df_train['Exited']==1]

[len(df_train_0['Exited']),len(df_train_1['Exited'])]
```

```python
df_train_ens1 = pd.concat([df_train_0[0:2122],df_train_1],axis='rows')
df_train_ens2 = pd.concat([df_train_0[2123:4245],df_train_1],axis='rows')
df_train_ens3 = pd.concat([df_train_0[4246:6368],df_train_1],axis='rows')
```

```python
y_ens1 = ann_model(df_train_ens1.drop(columns = 'Exited'),x_test,df_train_ens1['Exited'],y_test)
y_ens2 = ann_model(df_train_ens2.drop(columns = 'Exited'),x_test,df_train_ens2['Exited'],y_test)
y_ens3 = ann_model(df_train_ens3.drop(columns = 'Exited'),x_test,df_train_ens3['Exited'],y_test)
```

```python
y_pred_ens = max(y_ens1,y_ens2,y_ens3)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred_ens))
```
